```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-4,4,20)
f1 = -2*x-1
f2=np.sin(x)
# gera dois gr√°ficos simultaneamente
plt.plot(x,f1,x,f2,linewidth = 3)

plt.grid(True)
```

```python
import numpy as np
x = np.array([i for i in range(-3,4,1)]).reshape((1,7))
y=np.sin(x)+2*x+1
print(x)
print(np.round(y,2))
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-4,4,20)
f1 = np.exp(x)
f2=-x+3
# gera dois gr√°ficos simultaneamente
plt.plot(x,f1,x,f2,linewidth = 3)

plt.grid(True)
```

```python
import numpy as np
x = np.array([i for i in range(-3,4,1)]).reshape((1,7))
y=x+np.exp(x)-3
print(x)
print(np.round(y,2))
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(1,6,20)
f1 = np.log(x)
f2=-x**2+2
# gera dois gr√°ficos simultaneamente
plt.plot(x,f1,x,f2,linewidth = 3)

plt.grid(True)
```

```python
import numpy as np
x = np.array([i for i in range(1,7,1)]).reshape((1,6))
y=x**2-np.log(x)-2
print(x)
print(np.round(y,2))
```
